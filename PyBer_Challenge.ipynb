# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "resources/city_data.csv"
ride_data_to_load = "resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type

# Create the city dataframe for each type of city.
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]
```

```python
# Get the number of rides for each city type, using groupby.
ride_counts_by_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
ride_counts_by_type
```

```python
# 2. Using the city_data.csv, get the total drivers for each city type, using groupby.
total_drivers_by_city = city_data_df.groupby("type").sum()["driver_count"]
total_drivers_by_city
```

```python
#  3. Get the total amount of fares for each city type, using groupby.
total_fares_by_type = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares_by_type
```

```python
#  4. Get the average fare per ride for each city type,
# by dividing the sum of all the fares by the total rides.
avg_fare_per_ride = total_fares_by_type / ride_counts_by_type
avg_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type,
# by dividing the sum of all the fares by the total drivers
avg_fare_per_driver = total_fares_by_type / total_drivers_by_city
avg_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame.
# ride_avg_fare = [urban_avg_fare, suburban_avg_fare, rural_avg_fare]
# driver_avg_fare = [urban_fare_per_driver, suburban_fare_per_driver, rural_fare_per_driver]

pyber_summary_df = pd.DataFrame({
            "Total Rides": ride_counts_by_type,
            "Total Drivers": total_drivers_by_city,
            "Total Fares": total_fares_by_type,
            "Average Fare per Ride": avg_fare_per_ride,
            "Average Fare per Driver": avg_fare_per_driver})
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,.0f}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,.0f}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:.2f}".format)
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 9. Using groupby() on the pyber_data_df to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
fares_by_date_and_type_df = pyber_data_df.groupby(["date", "type"]).sum()[["fare"]]
fares_by_date_and_type_df.head()
```

```python
# 10. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
fares_by_date_and_type_df = fares_by_date_and_type_df.reset_index()
fares_by_date_and_type_df.head()
```

```python
# 11. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
fares_by_date_and_type_pivot = fares_by_date_and_type_df.pivot (index="date", columns='type', values='fare')
fares_by_date_and_type_pivot
```

```python
# 12. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2018-01-01':'2018-04-29'.
fares_by_date_and_type_pivot.loc['2019-01-01':'2019-04-29', :]
```

```python
# 13. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
fares_by_date_and_type_pivot.index = pd.to_datetime(fares_by_date_and_type_pivot.index)
```

```python
# 14. Check that the datatype for the index is datetime using df.info()
fares_by_date_and_type_pivot.info()
```

```python
# 15. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
# Note: NaN values are automatically converted to zero.
sum_of_fares_per_week = fares_by_date_and_type_pivot.resample("W").sum()
sum_of_fares_per_week
```

```python
# 16. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function.
# Pivot table: sum_of_fares_per_week

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
```

```python
# Create the plot.
# plt.plot(sum_of_fares_per_week, marker="*", color="blue", linewidth=2, label='Boston')
plt.plot(sum_of_fares_per_week)
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the x limit.
plt.xlim(pd.Timestamp('2019-01-06'), pd.Timestamp('2019-05-05'))
# Create a title.
plt.title("Total Fare by City Type")
# Add a grid.
plt.grid(True)
# Add the legend.
plt.legend(['Rural', 'Suburban', 'Urban'])
# Save the figure.
plt.savefig("analysis/PyBer_fare_summary.png")
```
